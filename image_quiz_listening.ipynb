```python
from openai import OpenAI
from dotenv import load_dotenv
import os
import json
```

```python
load_dotenv()#.env파일에 저장된 환경변수를 불러오는 함수
api_key=os.getenv("OPENAI_API_KEY")
client=OpenAI(api_key=api_key)#api_key를 이용하여 OpenAI와 통신
```

```python
response = client.audio.speech.create(
    #Text-to-Speech 기능
    model="tts-1-hd",
    voice="alloy",
    input="Hello world! This is a TTS test.",
)

response.write_to_file("hello_world.mp3")

#오디오,이미지 등 셀 안에 직접 출력할 수 있는 모듈
import IPython.display as ipd

ipd.Audio("hello_world.mp3")
```

```python
with open("C:/llm01/gpt/images/image_quiz_eng.json", 'r', encoding='utf-8') as f:
    eng_dict=json.load(f)

eng_dict
```

```python
output_dir = "C:/llm01/gpt/audio"
voices = ['alloy','ash','coral','echo','fable','onyx','nova','sage','shimmer']

for q in eng_dict:
    no=q['no'] #문제 번호 가져오기
    quiz = q['eng'] #영어 텍스트 가져오기
    quiz = quiz.replace("-1", "-One.\t")
    quiz = quiz.replace("-2", "-Two.\t")
    quiz = quiz.replace("-3", "-Three.\t")
    quiz = quiz.replace("-4", "-Four.\t")
    #\t는 목소리 끊어 읽기 도와줌

    print(no,quiz)

    #다양한 목소리로 분배
    voice = voices[no%len(voices)]

    response=client.audio.speech.create(
        model="tts-1-hd",
        voice=voice,
        input=f'#{no}.{quiz}',
    )

    response.write_to_file(f"{output_dir}/{no}.mp3")
```
